```python
#normal viz

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/srini/Desktop/IV/UPI/banks.csv")
months = df['Month']
transaction_values = df['Value']

plt.figure(figsize=(10, 6))
plt.style.use('dark_background')
colors = ['yellow']
plt.plot(months, transaction_values, color=colors[0])

plt.ylabel('Month')
plt.xlabel('Transaction Value')
plt.title('Monthly Transaction Value')
plt.grid(True, linestyle='--', alpha=0.5) 
plt.legend(['Transaction Value'], loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

df = pd.read_csv("C:/Users/srini/Desktop/IV/UPI/banks.csv")

months = df['Month']
transaction_values = df['Value']

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Monthly Transaction Value Animation')
ax.set_xlabel('Month')
ax.set_ylabel('Value')
ax.set_xlim(months.iloc[0], months.iloc[-1])
ax.grid(True, linestyle='--', alpha=0.5)
line, = ax.plot([], [], color='yellow', label='Transaction Value')
ax.legend(loc='upper left')


def update(frame):
    
    x = months.iloc[:frame+1]
    y = transaction_values.iloc[:frame+1]
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=len(months), interval=500, blit=True)

plt.show()
```
